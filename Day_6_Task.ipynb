```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (you can use your real query result)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'total_revenue': [400, 500, 750, 750, 350],
    'order_volume': [2, 2, 3, 3, 1]
}

df = pd.DataFrame(data)

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for revenue
ax1.bar(df['Month'], df['total_revenue'], color='skyblue', label='Total Revenue ()')
ax1.set_ylabel('Total Revenue ()', color='skyblue')
ax1.set_xlabel('Month')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Line chart for order volume on a second y-axis
ax2 = ax1.twinx()
ax2.plot(df['Month'], df['order_volume'], color='orange', marker='o', label='Order Volume')
ax2.set_ylabel('Order Volume', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Monthly Sales Trend: Revenue & Order Volume (2023)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```
